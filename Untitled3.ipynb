```python
# === cube(x) ===
def cube(x):
    return np.power(x, 3)

# Tests
print(cube(3))                  # 27
print(cube(np.arange(1, 11)))   # [  1   8  27  64 125 216 343 512 729 1000]

# Apply to a 2Ã—4 matrix
matrix = np.arange(1, 9).reshape((2, 4))
print(cube(matrix))
```

```python
# === Robust loss: psi1(x) ===
def psi1(x):
    x = np.array(x)
    return np.where(x**2 > 1, 2 * np.abs(x) - 1, x**2)

z = np.array([-0.5, -5, 0.9, 9])
print(psi1(z))  # [ 0.25  9.    0.81 17.  ]
```

```python
# === Generalized robust loss: psi2(x, c=1) ===
def psi2(x, c=1):
    x = np.array(x)
    c = np.array(c)
    return np.where(x**2 > c**2, 2 * c * np.abs(x) - c**2, x**2)

# Tests
print(np.allclose(psi1(z), psi2(z, c=1)))       # True
print(psi2(z, c=[1,1,1,10]))                    # vectorized c
print(psi2(z, c=-1))                            # negative c still broadcasts
```

```python
# === psi3 with input validation ===
def psi3(x, c=1):
    if np.size(c) != 1 or c <= 0:
        raise ValueError("Parameter c must be a single positive number")
    x = np.array(x)
    return np.where(x**2 > c**2, 2 * c * np.abs(x) - c**2, x**2)

print(psi3(z))  # same as psi2(z,1)
```

```python
# === Closure demo: adder(y) ===
x = 7
def adder(y):
    # Returns x+y but does not modify global x
    return x + y

print(adder(1))  # 8
print(x)         # 7 (unchanged)
```

```python
# === circle_area(r) ===
def circle_area(r):
    return math.pi * np.array(r) ** 2

print(circle_area([1, 2, 3]))  # [ 3.14159265 12.56637061 28.27433388]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# First ensure data is loaded correctly
try:
    gmp = pd.read_csv("gmp.dat", delim_whitespace=True)
    gmp['pop'] = gmp['gmp'] / gmp['pcgmp']
    print("Data loaded successfully. First few rows:")
    print(gmp.head())
except Exception as e:
    print(f"Error loading data: {str(e)}")
    # Create dummy data for testing if real data fails
    gmp = pd.DataFrame({
        'gmp': np.random.uniform(1e5, 1e6, 100),
        'pcgmp': np.random.uniform(3e4, 1e5, 100)
    })
    gmp['pop'] = gmp['gmp'] / gmp['pcgmp']
    print("Using synthetic data for demonstration")

def estimate_scaling_exponent1(initial_a, verbose=False):
    # Clean data
    clean_data = gmp[np.isfinite(gmp['pcgmp']) & np.isfinite(gmp['pop']) & (gmp['pop'] > 0)].copy()
    if len(clean_data) == 0:
        raise ValueError("No valid data points after cleaning")

    pcgmp = clean_data['pcgmp'].values
    pop = clean_data['pop'].values

    # Gradient descent parameters
    a = float(initial_a)
    learning_rate = 1e-12  # Reduced from original step_scale
    max_iters = 500        # Increased iterations
    tolerance = 1e-6       # More precise stopping criterion

    # Track progress
    history = {'a': [], 'mse': [], 'deriv': []}

    for iteration in range(max_iters):
        # Current prediction
        predicted = 6611 * (pop ** a)

        # Calculate MSE and derivative
        residuals = pcgmp - predicted
        mse = np.mean(residuals**2)

        # Numerical derivative (using central differences for better stability)
        h = 1e-8
        mse_plus = np.mean((pcgmp - 6611 * (pop ** (a + h))) ** 2)
        mse_minus = np.mean((pcgmp - 6611 * (pop ** (a - h))) ** 2)
        deriv = (mse_plus - mse_minus) / (2 * h)

        # Update history
        history['a'].append(a)
        history['mse'].append(mse)
        history['deriv'].append(deriv)

        # Stopping condition
        if abs(deriv) < tolerance:
            if verbose:
                print(f"Converged after {iteration} iterations")
            break

        # Update parameter (gradient descent)
        a -= learning_rate * deriv

        # Optional: adaptive learning rate
        if iteration > 0 and history['mse'][-1] > history['mse'][-2]:
            learning_rate *= 0.9  # Reduce learning rate if MSE increased

    else:
        if verbose:
            print(f"Stopped after reaching max iterations ({max_iters})")

    # Visualization of optimization path
    if verbose:
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.plot(history['a'], label='Parameter a')
        plt.xlabel('Iteration')
        plt.ylabel('a value')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history['mse'], label='MSE')
        plt.xlabel('Iteration')
        plt.ylabel('Mean Squared Error')
        plt.legend()
        plt.tight_layout()
        plt.show()

    return {
        'a': a,
        'iterations': iteration + 1,
        'converged': abs(deriv) < tolerance,
        'final_mse': mse,
        'final_deriv': deriv,
        'history': history if verbose else None
    }

# Test with different initial values
print("\nTesting with initial a = 0.15:")
result = estimate_scaling_exponent1(0.15, verbose=True)
print(result)

print("\nTesting with initial a = 0.10:")
result = estimate_scaling_exponent1(0.10, verbose=True)
print(result)
```

```python
# === estimate_scaling_exponent2(a, y0=6611, ...) ===
def estimate_scaling_exponent2(a, y0=6611, max_iters=100, deriv_step=0.001,
                              step_scale=1e-12, stop_deriv=0.01):
    iteration = 0
    deriv = float('inf')
    while iteration < max_iters and abs(deriv) > stop_deriv:
        iteration += 1
        mse1 = np.mean((gmp['pcgmp'] - y0 * gmp['pop'] ** a) ** 2)
        mse2 = np.mean((gmp['pcgmp'] - y0 * gmp['pop'] ** (a + deriv_step)) ** 2)
        deriv = (mse2 - mse1) / deriv_step
        a -= step_scale * deriv
    return {'a': a, 'iterations': iteration, 'converged': iteration < max_iters}

print(estimate_scaling_exponent2(0.15))
```

```python
# === estimate_scaling_exponent3(a, ...) ===
def estimate_scaling_exponent3(a, y0=6611, response=None, predictor=None,
                              max_iters=100, deriv_step=0.001,
                              step_scale=1e-12, stop_deriv=0.01):
    response = response if response is not None else gmp['pcgmp']
    predictor = predictor if predictor is not None else gmp['pop']
    def mse(a_val):
        return np.mean((response - y0 * predictor ** a_val) ** 2)
    iteration = 0
    deriv = float('inf')
    while iteration < max_iters and abs(deriv) > stop_deriv:
        iteration += 1
        deriv = (mse(a + deriv_step) - mse(a)) / deriv_step
        a -= step_scale * deriv
    return {'a': a, 'iterations': iteration, 'converged': iteration < max_iters}

print(estimate_scaling_exponent3(0.15))
```

```python
# === estimate_scaling_exponent4(a, ...) ===
def estimate_scaling_exponent4(a, y0=6611, response=None, predictor=None,
                              max_iters=100, deriv_step=0.001,
                              step_scale=1e-12, stop_deriv=0.01):
    response = response if response is not None else gmp['pcgmp']
    predictor = predictor if predictor is not None else gmp['pop']
    def mse(a_val):
        return np.mean((response - y0 * predictor ** a_val) ** 2)
    iteration = 0
    deriv = float('inf')
    while iteration < max_iters and abs(deriv) > stop_deriv:
        iteration += 1
        deriv = (mse(a + deriv_step) - mse(a)) / deriv_step
        a -= step_scale * deriv
    return {'a': a, 'iterations': iteration, 'converged': iteration < max_iters}

print(estimate_scaling_exponent4(0.15))
```

```python
# === estimate_scaling_exponent5(a, ...) with for-loop ===
def estimate_scaling_exponent5(a, y0=6611, response=None, predictor=None,
                              max_iters=100, deriv_step=0.001,
                              step_scale=1e-12, stop_deriv=0.01):
    response = response if response is not None else gmp['pcgmp']
    predictor = predictor if predictor is not None else gmp['pop']
    def mse(a_val):
        return np.mean((response - y0 * predictor ** a_val) ** 2)
    for iteration in range(1, max_iters + 1):
        deriv = (mse(a + deriv_step) - mse(a)) / deriv_step
        a -= step_scale * deriv
        if abs(deriv) <= stop_deriv:
            break
    return {'a': a, 'iterations': iteration, 'converged': iteration < max_iters}

print(estimate_scaling_exponent5(0.15))
```

```python
# === predict_plm(model, newdata) & plotting helpers ===
def predict_plm(model, newdata):
    assert 'a' in model and 'y0' in model
    a, y0 = model['a'], model['y0']
    if not isinstance(a, (int, float)) or not isinstance(y0, (int, float)):
        raise ValueError("Model parameters must be numeric scalars")
    return y0 * np.array(newdata) ** a

def plot_plm1(plm, start, end):
    x = np.linspace(start, end, 200)
    y = predict_plm(plm, x)
    plt.plot(x, y)
    plt.show()

def plot_plm2(plm, **kwargs):
    start = kwargs.pop('from', 1)
    end   = kwargs.pop('to',   10)
    x = np.linspace(start, end, 200)
    y = predict_plm(plm, x)
    plt.plot(x, y, **kwargs)
    plt.show()

def plot_plm3(plm, start, end, n=101, **kwargs):
    x = np.linspace(start, end, n)
    y = predict_plm(plm, x)
    plt.plot(x, y, **kwargs)
    plt.show()

# Example:
plm_model = {'a': 0.1258, 'y0': 6611}
plot_plm1(plm_model, 1, 100)
```

```python
# === Recursion examples: factorial & Fibonacci ===
def my_factorial(n):
    if n == 1:
        return 1
    return n * my_factorial(n - 1)

def fib(n):
    if n == 0 or n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

print("5! =", my_factorial(5))
print("fib(10) =", fib(10))
```
