LINEAR REGREESSION AND PLOTTING

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

# Load the data
url = "http://www.stat.cmu.edu/~cshalizi/uADA/13/hw/01/calif_penn_2011.csv"
calif_penn = pd.read_csv(url)

# Filter Pennsylvania data
penn = calif_penn[calif_penn["STATEFP"] == 42].copy()

# 1. Check for missing/infinite values
print("Missing values before cleaning:")
print(penn[["Median_household_income", "Median_house_value"]].isnull().sum())

# 2. Remove rows with missing or infinite values
penn_clean = penn[
    np.isfinite(penn["Median_household_income"]) &
    np.isfinite(penn["Median_house_value"])
].copy()

# 3. Verify cleaning worked
print("\nMissing values after cleaning:")
print(penn_clean[["Median_household_income", "Median_house_value"]].isnull().sum())

# Prepare data for regression
X = penn_clean["Median_household_income"]
y = penn_clean["Median_house_value"]
X = sm.add_constant(X)  # Adds intercept term

# Fit linear regression
model = sm.OLS(y, X).fit()

# Print regression coefficients
print("\nRegression coefficients:")
print(model.params)
print("\nModel summary:")
print(model.summary())

# Get Allegheny county data (adjust range as needed)
allegheny_rows = range(23, min(425, len(penn_clean)))  # Safer range
allegheny_medinc = penn_clean.iloc[allegheny_rows]["Median_household_income"]
allegheny_values = penn_clean.iloc[allegheny_rows]["Median_house_value"]

# Calculate fitted values
allegheny_fitted = model.params["const"] + model.params["Median_household_income"] * allegheny_medinc

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(allegheny_fitted, allegheny_values, alpha=0.6)
plt.xlabel("Model-predicted median house values ($)")
plt.ylabel("Actual median house values ($)")
plt.title("Actual vs Predicted House Values in Allegheny County")

# Add 45-degree line for reference
max_val = max(allegheny_fitted.max(), allegheny_values.max()) * 1.1
plt.plot([0, max_val], [0, max_val], color="grey", linestyle="--")
plt.xlim(0, max_val)
plt.ylim(0, max_val)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

MATRIX OPERATIONS

```python
factory = np.array([[40, 60], [1, 3]])
print("factory @ factory.T:\n", factory @ factory.T)

six_sevens = np.full((2, 3), 7)
print("six_sevens:\n", six_sevens)

print("factory @ six_sevens:\n", factory @ six_sevens)
output = np.array([10, 20])
print("factory @ output:\n", factory @ output)
print("output @ factory:\n", output @ factory)

print("Transpose:\n", factory.T)
print("Determinant:", np.linalg.det(factory))
print("Diagonal:", np.diag(factory))
factory[np.diag_indices(2)] = [35, 4]
print("Modified diagonal:\n", factory)

print("Inverse:\n", np.linalg.inv(factory))
print("factory * inverse:\n", factory @ np.linalg.inv(factory))

available = np.array([1600, 70])
solution = np.linalg.solve(factory, available)
print("Solution to Ax = b:", solution)
print("Check Ax = b:", factory @ solution)
```

NAMED MATRIX WITH PANDAS

```python
import pandas as pd

factory_df = pd.DataFrame(factory, index=["labor", "steel"], columns=["cars", "trucks"])
available_series = pd.Series([1600, 70], index=["labor", "steel"])
output_series = pd.Series([20, 10], index=["trucks", "cars"])

aligned_output = output_series[factory_df.columns]
result = factory_df @ aligned_output
print(result)
print("All constraints met:", (result <= available_series).all())
```
