### 1. Write a program to plot a line graph representing the temperature readings over a week: 20, 22, 19, 23, 21, 24, 20

```python
import matplotlib.pyplot as plt

# Temperature data for each day of the week
temps = [20, 22, 19, 23, 21, 24, 20]
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

plt.figure(figsize=(8, 4))
plt.plot(days, temps, marker='o', linestyle='-', color='blue')
plt.title("Temperature over a Week")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 2. Write a program to generate the arithmetic sequence starting at 5 with a common difference of 3, for 8 terms.

```python
start = 5
diff = 3
n_terms = 8

# Generate the sequence
sequence = [start + i * diff for i in range(n_terms)]

print("Arithmetic sequence:", sequence)
```

### 3. Write a program to calculate the volume under the surface z = x^2 + y^2 over the square region 0<=x, y<=1

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create grid for x and y over [0, 1]
nx, ny = 50, 50
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot 3D surface
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_title('Surface: z = x^2 + y^2 over [0,1] x [0,1]')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
import numpy as np

volume_analytic = 2 / 3
print("Analytic volume =", volume_analytic)

# Numerical approximation (simple 2D Riemann sum)
nx, ny = 101, 101
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

dx = 1 / (nx - 1)
dy = 1 / (ny - 1)
volume_numeric = np.sum(Z) * dx * dy
print("Numeric volume (approx) =", volume_numeric)
```

### 4. Explain the differences between compiled and interpreted languages, and classify Python in this context

#### Compilation vs. Interpretation

> Compiled Languages (e.g., C, C++):

- The source code is translated once by a compiler into machine code or an intermediate binary (executables).

- At runtime, the CPU directly executes that machine code.

- Pros: Very fast execution; optimizations applied ahead of time; no runtime parsing.

- Cons: You must recompile for each platform/architecture; slower editâdebug cycle.

> Interpreted Languages (e.g., JavaScript, Ruby, Python):

- Source code (or bytecode) is parsed and executed lineâbyâline or statementâbyâstatement by an interpreter at runtime.

- Pros: Greater portability (need an interpreter), rapid prototyping, dynamic typing.

- Cons: Slower execution, since parsing/bytecode interpretation happens on the fly.

- Python is traditionally considered an interpreted language:

- Then, a virtual machine (interpreter) executes that bytecode lineâbyâline.

- Modern Python can also be just-in-time (JIT) compiled (e.g., PyPy) or converted to optimized native code (via tools like Cython), but by default, Pythonâs development model is interpreted.
